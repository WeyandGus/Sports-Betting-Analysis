###**Cleaning Quantitative Data**

```python
# Importing Initial Quantitative Data
import pandas as pd
df = pd.read_csv('514NFLDataQuantitative.csv') #note: this file is uploaded in the drive, just upload locally and you can import it
df.head()
```

```python
#df.info()
```

```python
#df.isna().sum()
```

```python
# Filling OppTO NA's with 0
df['OppTO'] = df['OppTO'].fillna(0)
```

```python
# Setting NA for Home to 1 (Raiders were home in that single Obs)
df['Home'] = df['Home'].fillna(1)
```

```python
# Looking at object type vars
#df.select_dtypes(include='object')
```

```python
# Replacing #DIV/0! in 3weekOppTO with 0 and converting to float type column
df['3weekOppTO'] = df['3weekOppTO'].str.replace('#DIV/0!', '0')
df['3weekOppTO'] = df['3weekOppTO'].astype(float)
```

```python
#df.info()
```

```python
# Dropping non pre-game stats: PointsFor, PointsAgainst, 1stD, TotYd, PassY, RushY, TO, Opp1stD, OppTotYd, OppPassY, OppRushY, OppTO, xpOffense, xpDefense, xpSpTms
df.drop(columns = ['PointsFor', 'PointsAgainst','1stD', 'TotYd', 'PassY', 'RushY', 'TO', 'Opp1stD', 'OppTotYd', 'OppPassY', 'OppRushY', 'OppTO', 'xpOffense', 'xpDefense', 'xpSpTms'], inplace = True)
#df.head()
```

```python
# Filtering out obs where week = 1
df = df[df['Week'] > 1]
#df.head()
```

```python
# Mapping W to 1 and L to 0 in TeamResult
df['TeamResult'] = df['TeamResult'].map({'W': 1, 'L': 0})
df.head()
```

```python
# Binary column for Home
df['Home'] = df['Home'].astype(int)
df.head()
```

###**Merging Unstructured Data with Structured Data**

```python
quant_df = df
sent_df = pd.read_csv("NFLSentimentData.csv")
sent_df.head()
```

```python
# Minnestoa Vikings = Wrong in Sent (Correct: Minnesota Vikings)
# Philladelphia Eagles = Wrong in Sent (Correct: Philadelphia Eagles)
# Cincinati Bengals = Wrong in sent (Correct: Cincinnati Bengals)
# Tennesee Titans = Wrong in sent (Correct: Tennessee Titans)
```

```python
# Changing team_name to Team in sent_df and filtering out Week 1 and Week 0s
sent_df = sent_df[sent_df['Week'] > 1]
sent_df['Team'] = sent_df['team_name']
sent_df.drop(columns = ['team_name', 'week_number'], inplace = True)
```

```python
# Merging to get sentiment score for Team
merged_team_df = pd.merge(quant_df, sent_df, on=['Week', 'Team'], how='left')
merged_team_df['mean_sentiment_team'] = merged_team_df['mean_sentiment']
merged_team_df.drop(columns = ['mean_sentiment'], inplace = True)
merged_team_df.head()
```

```python
# Setting sent_df to Opp now instead of team
sent_df['Opp'] = sent_df['Team']
sent_df.drop(columns = ['Team'], inplace = True)
sent_df.head()
```

```python
# Merging to get sentiment score for Opp
merged_df = pd.merge(merged_team_df, sent_df, on=['Week', 'Opp'], how='left')
merged_df['mean_sentiment_opp'] = merged_df['mean_sentiment']
merged_df.drop(columns = ['mean_sentiment'], inplace = True)
merged_df.head()
```
